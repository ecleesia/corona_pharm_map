```python
# 계양구_약국_데이터 엑셀 파일 추출

file_xl = pd.read_excel("계양구_약국_데이터.xlsx", encoding = "utf-8", index_col=0)
```

```python
# 위도 경도 기반하여 map에 표시하는 함수 생성
# popup은 이름과 전화번호 띄우도록
# popup에서 한글이 깨져서 html로 추출

def show_marker_map(geo_df):
    map = folium.Map(location = [geo_df["위도"].mean(), geo_df["경도"].mean()],
                    zoom_start=15, tiles = "Stamen Toner")
    
    for n in geo_df.index:
        pharm_name = geo_df.loc[n,"업체명"] + " - " + geo_df.loc[n, "전화번호"]
        folium.Marker([geo_df.loc[n, "위도"],
                       geo_df.loc[n, "경도"]],
                     popup = pharm_name).add_to(map)
        
    return map.save('pharm.html')
```

```python
show_marker_map(file_xl)
```
